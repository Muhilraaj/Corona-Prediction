```
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
```

```
#importing data
df=pd.read_excel("Train_dataset.xlsx",sheet_name="Diuresis_TS")
df.drop("people_ID",axis=1,inplace=True)
```

```
df=df.transpose()
df.head(100)
```

```
#building var model
model = VAR(df)
model_fitted = model.fit(5)
fr_cast_7=model_fitted.forecast(model_fitted.y,steps=1)
```

```
fr_cast_7
```

```
df7=pd.DataFrame()
df7.insert(0,"Today",df.iloc[0,:],True)
df7.insert(1,"27th",np.transpose(fr_cast_7),True)
df7.head(100)
```

```
c=0
def adfuller_test(series, signif=0.05, name='', verbose=False):
    global c
    r = adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue'] 
    def adjust(val, length= 6): return str(val).ljust(length)
    if p_value <= signif:
        c+=1
        return 1
    else:
        return 0
```

```
dt=pd.read_excel("Test_dataset.xlsx")
```

```
n=len(dt)
#print(min(abs(df7['Today']-dt['Diuresis'][0])))
s=0
ds7=dt['Diuresis'].copy()
for i in range(0,n):
  s=abs(df7['Today']-dt['Diuresis'][i])
  j=np.where(s==min(s))
  st=[]
  for ii in j:
    st.append(df7['27th'][ii])
  dt['Diuresis'][i]=np.median(st)
ds7
```
