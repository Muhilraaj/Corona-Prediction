```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error 
from scipy import stats
```

```
#importing preprocesses data
df=pd.read_csv("Train.csv")
df.columns
```

```
#outlier detection and removal
z = np.abs(stats.zscore(df))
print(z)
```

```
df=df[(z < 3).all(axis=1)]
df.shape
```

```
#data splitting
y=df['Infect_Prob']
X=df.drop('Infect_Prob',axis=1)
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```
def poly(X):
  p=PolynomialFeatures(degree=1)
  X=p.fit_transform(X)
  return X
```

```
X_train=poly(X_train)
X_test=poly(X_test)
X=poly(X)
X=pd.DataFrame(X)
X.head(10)
```

```
#model train
model=LinearRegression()
#model=Lasso(alpha=3.0,max_iter=3000)
model.fit(X,y)
y_pred_test=model.predict(X_test)
y_pred_train=model.predict(X_train)
```

```
#accuracy check
print(mean_squared_error(y_pred_train,y_train))
print(mean_squared_error(y_pred_test,y_test))
```

```
#importing test data for 20th and 27th
dft_20=pd.read_csv("Test_20th.csv")
dft_27=pd.read_csv("Test_27th.csv")
pid=pd.read_excel("Test_dataset.xlsx")
dft_20=poly(dft_20)
dft_27=poly(dft_27)
```

```
#making prediction from existing model
ft_20=model.predict(dft_20)
ft_27=model.predict(dft_27)
```

```
fi_20=pd.DataFrame()
fi_20.insert(0,'people_ID',pid['people_ID'],True)
fi_20.insert(1,'Infect_Prob',ft_20,True)
fi_27=pd.DataFrame()
fi_27.insert(0,'people_ID',pid['people_ID'],True)
fi_27.insert(1,'Infect_Prob',ft_27,True)
```
